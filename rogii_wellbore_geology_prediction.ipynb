```python
import os
import pandas as pd
from tqdm import tqdm

base_path = '/kaggle/input/competitions/rogii-wellbore-geology-prediction'

def load_well_data(folder_name):
    folder_path = os.path.join(base_path, folder_name)
    horizontal_wells = []
    type_wells = []
    
    for filename in tqdm(os.listdir(folder_path)):
        file_path = os.path.join(folder_path, filename)
        well_id = filename.split('__')[0] 
        
        try:
            df = pd.read_csv(file_path)
            df['well_id'] = well_id 
            
            if 'horizontal_well' in filename:
                horizontal_wells.append(df)
            elif 'typewell' in filename:
                type_wells.append(df)
        except Exception as e:
            print(f"Error reading {filename}: {e}")
        
    all_horizontal = pd.concat(horizontal_wells, ignore_index=True) if horizontal_wells else pd.DataFrame()
    all_type = pd.concat(type_wells, ignore_index=True) if type_wells else pd.DataFrame()
    
    return all_horizontal, all_type

print("Loading Train Data...")
train_horizontal, train_type = load_well_data('train')

print("\nLoading Test Data...")
test_horizontal, test_type = load_well_data('test')

print("\n--- Summary ---")
print(f"Train Horizontal Shape: {train_horizontal.shape}")
print(f"Test Horizontal Shape: {test_horizontal.shape}")
```

```python
import pandas as pd

features = ['MD', 'X', 'Y', 'Z', 'GR', 'TVT_input']
target = 'TVT'

print("======= TRAIN DATA CHECK =======")
print(f"Total Rows in Train: {len(train_horizontal)}\n")
train_info = pd.DataFrame({
    'Data Type': train_horizontal[features + [target]].dtypes,
    'Missing Values': train_horizontal[features + [target]].isnull().sum(),
    'Missing Percentage (%)': (train_horizontal[features + [target]].isnull().sum() / len(train_horizontal)) * 100
})
print(train_info)

print("\n======= TEST DATA CHECK =======")
print(f"Total Rows in Test: {len(test_horizontal)}\n")
test_info = pd.DataFrame({
    'Data Type': test_horizontal[features].dtypes,
    'Missing Values': test_horizontal[features].isnull().sum(),
    'Missing Percentage (%)': (test_horizontal[features].isnull().sum() / len(test_horizontal)) * 100
})
print(test_info)
```

```python
import pandas as pd
import numpy as np

features = ['MD', 'X', 'Y', 'Z', 'GR', 'TVT_input']
target = 'TVT'

print("Starting Advanced Group-based Data Cleaning...")

train_horizontal = train_horizontal.dropna(subset=[target]).reset_index(drop=True)

for col in features:
    train_horizontal[col] = train_horizontal[col].replace([np.inf, -np.inf], np.nan)
    test_horizontal[col] = test_horizontal[col].replace([np.inf, -np.inf], np.nan)

print("Imputing missing values using forward-fill and backward-fill per well...")

train_horizontal[features] = train_horizontal.groupby('well_id')[features].ffill()
train_horizontal[features] = train_horizontal.groupby('well_id')[features].bfill()

test_horizontal[features] = test_horizontal.groupby('well_id')[features].ffill()
test_horizontal[features] = test_horizontal.groupby('well_id')[features].bfill()

print("Applying global median for any remaining isolated missing values...")
for col in features:
    global_median = train_horizontal[col].median()
    train_horizontal[col] = train_horizontal[col].fillna(global_median)
    test_horizontal[col] = test_horizontal[col].fillna(global_median)
print("\nAll Missing Values Successfully Handled!")
print(f"Remaining Nulls in Train: {train_horizontal[features].isnull().sum().sum()}")
print(f"Remaining Nulls in Test: {test_horizontal[features].isnull().sum().sum()}")
```

```python
import lightgbm as lgb
from sklearn.model_selection import KFold
import numpy as np
import pandas as pd
import gc
```

```python
X = train_horizontal[features]
y = train_horizontal[target]
X_test = test_horizontal[features]

kf = KFold(n_splits=10, shuffle=True, random_state=42)
test_predictions = np.zeros(len(X_test))

print("Fitting Hyperparameter-Tuned Model on Cleaned Data...")

for fold, (train_idx, val_idx) in enumerate(kf.split(X, y)):
    X_train, y_train = X.iloc[train_idx], y.iloc[train_idx]
    X_val, y_val = X.iloc[val_idx], y.iloc[val_idx]
    
    model = lgb.LGBMRegressor(
        n_estimators=2500,
        learning_rate=0.01,
        max_depth=10,
        num_leaves=127,
        min_child_samples=250,
        min_split_gain=0.01,
        reg_alpha=0.1,
        reg_lambda=1.0,
        subsample=0.75,
        colsample_bytree=1.0,
        random_state=42,
        n_jobs=-1
    )
    
    model.fit(
        X_train, y_train,
        eval_set=[(X_val, y_val)],
        callbacks=[lgb.early_stopping(stopping_rounds=50, verbose=False)]
    )
    
    test_predictions += model.predict(X_test) / kf.n_splits
    del X_train, y_train, X_val, y_val, model
    gc.collect()
    print(f"Fold {fold + 1} Fit Successfully and Memory Cleared.")

print("\nMapping Predictions to Kaggle's Official Sample Submission...")
sample_sub = pd.read_csv('/kaggle/input/competitions/rogii-wellbore-geology-prediction/sample_submission.csv')

test_horizontal['id'] = test_horizontal['well_id'].astype(str) + "_" + test_horizontal['MD'].round().astype(int).astype(str)
test_horizontal['tvt_pred'] = test_predictions

pred_map = test_horizontal.groupby('id')['tvt_pred'].mean().to_dict()

sample_sub['tvt'] = sample_sub['id'].map(pred_map)

global_median = y.median()
sample_sub['tvt'] = sample_sub['tvt'].fillna(global_median)

sample_sub.to_csv('submission.csv', index=False)
print("\n--- Process Finished Successfully! ---")
print(f"Submission Shape: {sample_sub.shape}")
print(sample_sub.head())
```
